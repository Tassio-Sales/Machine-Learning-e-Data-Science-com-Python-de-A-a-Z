Quest√£o 1: Ler dois n√∫meros inteiros, executar e mostrar o resultado das seguintes opera√ß√µes: adi√ß√£o, subtra√ß√£o, multiplica√ß√£o e divis√£o



```python
def ler_inteiro(mensagem: str) -> int:
    """L√™ um n√∫mero inteiro com tratamento de erro."""
    while True:
        try:
            return int(input(mensagem).strip())
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida. Por favor, insira um n√∫mero inteiro.\n")


def calculadora_de_inteiros() -> None:
    """Executa opera√ß√µes matem√°ticas b√°sicas entre dois n√∫meros inteiros."""
    print("=== Calculadora de Opera√ß√µes Matem√°ticas entre Inteiros ===\n")

    n1 = ler_inteiro("Digite o primeiro n√∫mero inteiro: ")
    n2 = ler_inteiro("Digite o segundo n√∫mero inteiro: ")

    # Opera√ß√µes b√°sicas
    soma = n1 + n2
    subtracao = n1 - n2
    multiplicacao = n1 * n2
    potencia = n1 ** n2

    # Divis√µes com tratamento
    if n2 != 0:
        divisao = n1 / n2
        divisao_inteira = n1 // n2
        resto_divisao = n1 % n2
    else:
        divisao = divisao_inteira = resto_divisao = "‚ùå Divis√£o por zero."

     # Exibi√ß√£o dos resultados
    print("\nüìä Resultados das Opera√ß√µes:\n" + "-" * 40)
    print(f"‚ûï Soma               : {soma}")
    print(f"‚ûñ Subtra√ß√£o          : {subtracao}")
    print(f"‚úñÔ∏è  Multiplica√ß√£o     : {multiplicacao}")
    print(f"^  Pot√™ncia           : {potencia}")
    print(f"‚ûó Divis√£o            : {divisao}")
    print(f"// Divis√£o Inteira    : {divisao_inteira}")
    print(f"%  Resto da Divis√£o   : {resto_divisao}")
    print("-" * 40)
    print("\n‚úÖ C√°lculos conclu√≠dos com sucesso!")

if __name__ == "__main__":
    try:
        calculadora_de_inteiros()
    except KeyboardInterrupt:
        print("\n\n‚ùå Execu√ß√£o interrompida pelo usu√°rio.")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
```

Quest√£o 2: Efetuar o c√°lculo da quantidade de litros de combust√≠vel gasto em uma viagem, utilizando um autom√≥vel que faz 12 Km por litro. Para obter o c√°lculo, o usu√°rio deve fornecer o tempo gasto na viagem e a velocidade m√©dia durante ela. Desta forma, ser√° poss√≠vel obter a dist√¢ncia percorrida com a f√≥rmula DISTANCIA = TEMPO * VELOCIDADE. Tendo o valor da dist√¢ncia, basta calcular a quantidade de litros de combust√≠vel utilizada na viagem, com a f√≥rmula: LITROS_USADOS = DISTANCIA / 12. O programa deve apresentar os valores da velocidade m√©dia, tempo gasto na viagem, a dist√¢ncia percorrida e a quantidade de litros utilizada na viagem

```python
def ler_float_positivo(mensagem: str) -> float:
    """L√™ um n√∫mero float positivo com tratamento de erro e valida√ß√£o."""
    while True:
        try:
            valor = float(input(mensagem).strip())
            if valor <= 0:
                print("‚ö†Ô∏è O valor deve ser maior que zero.\n")
                continue
            return valor
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida. Por favor, insira um valor num√©rico.\n")

def calculadora_de_viagem() -> None:
    """Calcula a dist√¢ncia percorrida e o consumo de combust√≠vel em uma viagem."""
    print("=== Calculadora de Consumo de Viagem ===\n")
    
    tempo = ler_float_positivo("‚è±Ô∏è  Digite o tempo de viagem (em horas): ")
    velocidade_media = ler_float_positivo("üí® Digite a velocidade m√©dia (em km/h): ")

    # C√°lculos
    distancia = tempo * velocidade_media
    litros_usados = distancia / 12  # Consumo m√©dio de 12 km/l

    # Exibi√ß√£o dos resultados
    print("\nüìä Resultados da Viagem:")
    print("-" * 45)
    print(f"üïì Tempo gasto: {tempo:.2f} horas")
    print(f"üöó Velocidade m√©dia: {velocidade_media:.2f} km/h")
    print(f"üõ£Ô∏è  Dist√¢ncia percorrida: {distancia:.2f} km")
    print(f"‚õΩ Combust√≠vel utilizado: {litros_usados:.2f} litros")
    print("-" * 45)
    print("\n‚úÖ C√°lculos conclu√≠dos com sucesso!")

if __name__ == "__main__":
    try:
        calculadora_de_viagem()
    except KeyboardInterrupt:
        print("\n\n‚ùå Execu√ß√£o interrompida pelo usu√°rio.")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
```
