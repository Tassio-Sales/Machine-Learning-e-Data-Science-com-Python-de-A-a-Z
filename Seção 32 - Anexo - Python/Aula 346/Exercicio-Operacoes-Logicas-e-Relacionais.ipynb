Questão 1: Leia a idade do usuário e classifique-o em:
- Criança – 0 a 12 anos
- Adolescente – 13 a 17 anos
- Adulto – acima de 18 anos
-Se o usuário digitar um número negativo, mostrar a mensagem que a idade é inválida

```python
def ler_idade(mensagem: str) -> int:
    """Lê e valida um valor inteiro não negativo referente à idade."""
    while True:
        try:
            idade = int(input(mensagem).strip())
            if idade < 0:
                print("⚠️ Idade inválida. A idade não pode ser negativa.\n")
                continue
            return idade
        except ValueError:
            print("⚠️ Entrada inválida. Por favor, insira um número inteiro.\n")

def classificar_idade_usuario() -> None:
    """Lê a idade do usuário e exibe a classificação correspondente."""
    print("=== Classificador de Idade ===\n")
    
    idade = ler_idade("Digite sua idade: ")

    if idade <= 12:
        classificacao = "Criança"
    elif idade <= 17:
        classificacao = "Adolescente"
    else:
        classificacao = "Adulto"

    print(f"\nVocê é classificado como: {classificacao}")

if __name__ == "__main__":
    try:
        classificar_idade_usuario()
    except KeyboardInterrupt:
        print("\n\n❌ Programa encerrado pelo usuário.")
    except Exception as e:
        print(f"\n⚠️ Ocorreu um erro inesperado: {e}")
```

Questão 2: Calcular a média de um aluno que cursou a disciplina de Programação I, a partir da leitura das notas M1, M2 e M3; passando por um cálculo da média aritmética. Após a média calculada, devemos anunciar se o aluno foi aprovado, reprovado ou pegou exame
- Se a média estiver entre 0.0 e 4.0, o aluno está reprovado
- Se a média estiver entre 4.1 e 6.0, o aluno pegou exame
- Se a média for maior do que 6.0, o aluno está aprovado
- Se o aluno pegou exame, deve ser lida a nota do exame. Se a nota do exame for maior do que 6.0, está aprovado, senão; está reprovado

```python
def ler_nota(mensagem: str, minimo: float = 0.0, maximo: float = 10.0) -> float:
    """
    Lê e valida uma nota (float) no intervalo [minimo, maximo].
    Aceita vírgula como separador decimal (ex: '7,5').
    """
    while True:
        try:
            raw = input(mensagem).strip()
            # permite entrada com vírgula
            raw = raw.replace(",", ".")
            nota = float(raw)
            if nota < minimo or nota > maximo:
                print(f"⚠️ Nota inválida. Informe um valor entre {minimo} e {maximo}.\n")
                continue
            return nota
        except ValueError:
            print("⚠️ Entrada inválida. Por favor, insira um número válido (ex: 7.5 ou 7,5).\n")

def calcular_media() -> None:
    """Calcula a média aritmética de M1, M2 e M3 e decide situação do aluno."""
    print("\n=== Calculadora de Média (Programação I) ===\n")
    try:
        m1 = ler_nota("Digite a nota M1: ")
        m2 = ler_nota("Digite a nota M2: ")
        m3 = ler_nota("Digite a nota M3: ")
    except KeyboardInterrupt:
        print("\n\n❌ Entrada cancelada pelo usuário.")
        return

    media = (m1 + m2 + m3) / 3
    print(f"\nMédia do aluno: {media:.2f}")

    # Classificação sem lacunas entre intervalos
    if media <= 4.0:
        print("Status: Reprovado")
    elif media <= 6.0:  
        print("Status: Pegou Exame")
        try:
            nota_exame = ler_nota("Digite a nota do exame: ")
        except KeyboardInterrupt:
            print("\n\n❌ Entrada cancelada pelo usuário.")
            return

        # regra pedida: exame > 6.0 => aprovado; caso contrário reprovado
        if nota_exame > 6.0:
            print(f"Resultado do exame: {nota_exame:.2f} -> Aprovado")
        else:
            print(f"Resultado do exame: {nota_exame:.2f} -> Reprovado")
    else:  # media > 6.0
        print("Status: Aprovado")


if __name__ == "__main__":
    try:
        calcular_media()
    except KeyboardInterrupt:
        print("\n\n❌ Programa encerrado pelo usuário.")
    except Exception as e:
        print(f"\n⚠️ Ocorreu um erro inesperado: {e}")
```
