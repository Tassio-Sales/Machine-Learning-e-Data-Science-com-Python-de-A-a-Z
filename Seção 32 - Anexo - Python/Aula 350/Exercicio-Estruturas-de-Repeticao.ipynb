Quest√£o 1: Ler 5 notas e informar a m√©dia

```python
def ler_nota(mensagem: str, minimo: float = 0.0, maximo: float = 10.0) -> float:
    """
    L√™ e valida uma nota (float) no intervalo [minimo, maximo].
    Aceita v√≠rgula como separador decimal (ex: '7,5').
    """
    while True:
        try:
            raw = input(mensagem).strip()            
            raw = raw.replace(",", ".")
            nota = float(raw)
            if nota < minimo or nota > maximo:
                print(f"‚ö†Ô∏è Nota inv√°lida. Informe um valor entre {minimo} e {maximo}.\n")
                continue
            return nota
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida. Por favor, insira um n√∫mero v√°lido (ex: 7.5 ou 7,5).\n")

def calcular_media() -> None:
    """Calcula a m√©dia aritm√©tica entre as 5 notas do aluno, com repeti√ß√£o controlada."""
    print("\n=== Calculadora de M√©dia ===\n")

    while True:
        try:
            notas = []
            for i in range(1, 6):
                nota = ler_nota(f"Digite a nota {i}: ")
                notas.append(nota)

            media = sum(notas) / len(notas)
            print(f"\nüìä A m√©dia do aluno √©: {media:.2f}\n")

            # Pergunta se deseja calcular novamente
            while True:
                repetir = input("Deseja calcular outra m√©dia? (S/N): ").strip().lower()
                if repetir == "s":
                    print("\nüîÑ Recome√ßando o c√°lculo...\n")
                    break
                elif repetir == "n":
                    print("\n‚úÖ Programa encerrado. At√© a pr√≥xima!\n")
                    return
                else:
                    print("‚ö†Ô∏è Op√ß√£o inv√°lida. Digite apenas 'S' para sim ou 'N' para n√£o.\n")

        except KeyboardInterrupt:
            print("\n\n‚ùå Entrada cancelada pelo usu√°rio.")
            return
        except Exception as e:
            print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
            return


if __name__ == "__main__":
    try:
        calcular_media()
    except KeyboardInterrupt:
        print("\n\n‚ùå Programa encerrado pelo usu√°rio.")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
```

Quest√£o 2: Imprimir a tabuada do n√∫mero 3 (3 x 1 = 3 ; 3 x 10 = 30)

```python
def ler_inteiro(mensagem: str) -> int:
    """L√™ um n√∫mero inteiro com valida√ß√£o e tratamento de erro."""
    while True:
        try:
            return int(input(mensagem))
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida. Por favor, insira um n√∫mero inteiro.\n")
           
def tabuada() -> None:
    """Imprime a tabuada de um n√∫mero informado pelo usu√°rio."""
    print("\n=== Gerador de Tabuada ===\n")

    while True:
        try:
            numero = ler_inteiro("Digite o n√∫mero para ver a tabuada: ")

            print(f"\nüìò Tabuada do {numero}")
            print("-" * 20)

            for i in range(1, 11):
                print(f"{numero:>2} x {i:>2} = {numero * i:>3}")

            print("-" * 20)

            while True:
                repetir = input("\nDeseja ver outra tabuada? (S/N): ").strip().lower()
                if repetir == "s":
                    print()
                    break
                elif repetir == "n":
                    print("\n‚úÖ Programa encerrado. At√© a pr√≥xima!\n")
                    return
                else:
                    print("‚ö†Ô∏è Op√ß√£o inv√°lida. Digite apenas 'S' ou 'N'.")

        except KeyboardInterrupt:
            print("\n\n‚ùå Entrada cancelada pelo usu√°rio.")
            return
        except Exception as e:
            print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
            return


if __name__ == "__main__":
    try:
        tabuada()
    except KeyboardInterrupt:
        print("\n\n‚ùå Programa encerrado pelo usu√°rio.")
    except Exception as e:
        print(f"\n‚ö†Ô∏è Ocorreu um erro inesperado: {e}")
```
